```python
import sys, os
sys.path.append(os.path.abspath(".."))

from core.ollama_client import chat_with_ollama
```

```python
system_prompt = """
You are a helpful assistant that can explain code.
Explain the code in a way that is easy to understand.
Use markdown to format your response and use code blocks to format the code with colors if needed.
"""

question_user_prompt = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
def explain_code():
  stream = chat_with_ollama(
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": question_user_prompt}
      ],
  )
  return stream
```
